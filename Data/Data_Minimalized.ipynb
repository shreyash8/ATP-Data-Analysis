This Notebook takes CSV file of ATP Data and Divides this Data into several CSV files.
This several CSV files will help us to create a perfect minimalized Database using MySql.

The minimalized CSV files are created using Database Concepts.

```python
import csv
import pandas as pd
```

```python
# Reading Data using Pandas 
data = pd.read_csv("ATP.csv")
```

```python
# We are dropping data before year 2000. (As Data after 2000 is very accurate as compared to before 2000)
arr = []
for i in range(110457):
    arr.append(i)
```

```python
data = data.drop(arr)
data

# Visualization of our Data
```

```python
# Printing All the features of our Data
i = 0
for col in data.columns: 
    print(i, col)
    i = i+1
```

```python
tt2 = data[['tourney_id', 'match_num','round', 'winner_id', 'winner_rank', 'winner_rank_points','winner_entry', 'winner_seed', 'loser_id', 'loser_rank', 'loser_rank_points', 'loser_entry', 'loser_seed' ]]
```

```python
tt2

# Creating our 1st minimal Data. (We have dropped some features here suitably)
```

```python
# Creating CSV file

tt2.to_csv('match_info.csv', index = 'false')
```

```python
# Creating 2nd Table of our Data

tt3 = data[['tourney_id', 'match_num', 'winner_name', 'winner_age', 'winner_ioc', 'winner_ht' ,'winner_hand', 'loser_name', 'loser_age', 'loser_ioc', 'loser_ht', 'loser_hand']]
```

```python
tt3
```

```python
# Converting to CSV

tt3.to_csv('extra_info.csv', index = 'false')
```

```python
# Similarly as above.

tt4 = data[['tourney_id', 'match_num', 'minutes', 'score', 'w_ace', 'w_df', 'w_svpt', 'w_1stIn', 'w_1stWon', 'w_2ndWon','w_SvGms','w_bpSaved','w_bpFaced', 'l_ace','l_df','l_svpt','l_1stIn','l_1stWon','l_2ndWon','l_SvGms','l_bpSaved','l_bpFaced']]
```
